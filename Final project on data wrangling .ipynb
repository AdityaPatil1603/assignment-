# 1. Import the dataset using Pandas from above mentioned url

```python
# Importing the data from website 
import pandas as pd 
import numpy as np 
import datetime 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
url='https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv'
```

```python
df=pd.read_csv(url)
df.head()
```

# 2. High Level Data Understanding:
   

```python
# a.Finding no. of rows and columns in dataframe 
print('no of rows in dataset :',len(df.axes[0]))
print('no of columns in dataset:',len(df.axes[1]))
```

```python
# b. Data types of columns.
df.dtypes
```

```python
#c.Info of data in dataframe.
df.info()
```

```python
# c.describe of data in dataframe.
df.describe()
```

# 3. Low Level Data Understanding :

```python
#a. Find count of unique values in location column
df['location'].nunique()
```

```python
# b.Find which continent has maximum frequency using values counts.
df['continent'].value_counts().head(1)
```

```python
# c. Find maximum & mean value in 'total_cases'.
print('maximum value of total casese:',df['total_cases'].max())
print('mean value of total cases:',df['total_cases'].mean())
```

```python
# d.Find 25%,50% & 75% quartile value in 'total_deaths'.
print('25% quartile value in total_deaths:',df['total_deaths'].quantile(q=0.25))
print('50% quartile value in total_deaths:',df['total_deaths'].quantile(q=0.50))
print('75% quartile value in total_deaths:',df['total_deaths'].quantile(q=0.75))
```

```python
# e.Find which continent has maximum'human_development_index'.
df.groupby('continent')['human_development_index'].max().sort_values(ascending=False).head(1)
```

```python
# f.Find which continent has minimum 'gdp_per_capita'.
df.groupby('continent')['gdp_per_capita'].min().sort_values(ascending=True).head(1)
```

# 4.Filter the dataframe with only this columns['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.

```python
#. Filter the dataframe with only this columns['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.
df.loc[:, df.columns.isin(['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] )]
```

# 5. Data Cleaning

```python
# a.Remove all duplicates observations
df.drop_duplicates(inplace=True)
```

```python
# b.Find missing values in all columns
df.isna()
```

```python
# c.Remove all observations where continent column value is missing
df.dropna(subset=['continent'])
```

```python
# d.Fill all missing values with 0.
df.fillna(0, inplace=True)
```

# 6. Date time format :

```python
# a. Convert date column in datetime format using pandas.to_datetime
df['date']=pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
# b. Create new column month after extracting month data from date column.
df['month'] = pd.DatetimeIndex(df['date']).month
df.head()
```

# 7. Data Aggregation:


```python
#a. Find max value in all columns using groupby function on 'continent' column
df.groupby('continent').max().reset_index()
```

```python
# b. Store the result in a new dataframe named 'df_groupby'.
df_groupby = df.groupby('continent').max().reset_index()
df_groupby.head(10)
```

# 8. Feature Engineering :

```python
# a. Create a new feature 'total_deaths_to_total_cases' by ratio of
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
df_groupby.to_csv(url, index=False)
df_groupby.head(30)
```

```python
# a. Perform Univariate analysis on 'gdp_per_capita' column by plotting\histogram using seaborn dist plot.
plt.hist(df_groupby['gdp_per_capita'])
```

```python
# b.Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
sns.scatterplot(x='gdp_per_capita', y='total_cases', data=df_groupby, color='blue')
plt.title('Total Cases vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Total Cases')
```

```python
#c. Plot Pairplot on df_groupby dataset.
pair_plot=sns.pairplot(df_groupby.sample(10))
print(pair_plot)
```

```python
#d. Plot a bar plot of 'continent' column with 'total_cases' .Tip : using kind='bar' in seaborn catplot
sns.catplot('continent','total_cases', data=df, kind='bar')
```
